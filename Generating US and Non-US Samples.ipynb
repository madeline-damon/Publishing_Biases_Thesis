# Removing emails that are not the first author's

Just like the author address data, the email data needed to send survey also contains all author emails, not just the first author. I need to grab only the first author emails.

```python
import re
email_regex=r"[a-z0-9!#$%&'*+/=?^_`{|}~-]+(?:\.[a-z0-9!#$%&'*+/=?^_`{|}~-]+)*@(?:[a-z0-9](?:[a-z0-9-]*[a-z0-9])?\.)+[a-z0-9](?:[a-z0-9-]*[a-z0-9])?;?"
#This captures single emails but not when there are multiple
#r'^[a-zA-Z0-9]+(?:\.[a-zA-Z0-9]+)*@[a-zA-Z0-9]+(?:\.[a-zA-Z0-9]+)*$'
```

```python
US_data=pd.read_csv('US_Survey_Sample.csv')
NonUS_data=pd.read_csv('NonUS_Survey_Sample.csv')
```

## United States Sample

```python
US_email_list=list(US_data['Email Addresses'])
```

```python
US_email_strings=[]

for item in US_email_list:  #Make everything in the column  a string of text
    string=str(item)
    US_email_strings.append(string)
```

```python
US_email_strings
```

```python
#finding regex matches
US_match_list=list()

for email in US_email_strings:
    match=re.findall(email_regex,email)
    US_match_list.append(match)
```

```python
US_match_list
```

```python
#Cleaning the list of regex matches to remove semi colons for multiple entries
list_of_matches=list()

for matches in US_match_list:
    in_list=matches
    out_list= [re.sub(';', '', item) for item in in_list]
    
    list_of_matches.append(out_list)
```

```python
len(list_of_matches)
```

```python
list_of_matches
```

```python
first_author_email=list()
for emails in list_of_matches:
    
     if len(emails)==1:
        first_author_email.append(emails)   
     
     elif len(emails)==0:
        first_author_email.append('NA')
    
     else:
        first_author_email.append(emails[0])
```

```python
first_author_email
```

```python
US_data['First_A_Email']=first_author_email
```

```python
US_data
```

```python
US_data.to_csv('US_Survey_Sample.csv')
```
